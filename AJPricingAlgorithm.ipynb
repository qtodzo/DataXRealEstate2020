```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as mse
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
```

```python
file_path='./Assessor_Historical_Secured_Property_Tax_Rolls.csv'
data=pd.read_csv(file_path)
data#.head()
```

```python
cols = list(data.columns)#, list(data.index)
cols#, indices[:10]
#len(cols) == 53
#0
#5 volume number
#6 Use Code
#8 Property Class Code
#10 year built
#[11-15] number of br, bedr, room, stories, units
#17 const type
#[18-22] Lot areas (depth, frontage, prop area, basement area, lot area)
#[32-35] Values
```

```python
##Delete irrelevant end features
data = data.iloc[:, :-6]
#data.head()
```

```python
sub = data.iloc[:, [0,5,6,8,10,11,12,13,14,15,17,18,19,20,21,22,32,33,34,35]]
sub.head()
```

```python
#SINGLE RESIDENTIAL. UNINTERESTED IN COMMERCIAL, GOVT, ETC. THIS HELPS REDUCE LENGTH OF DATASET
```

```python
sub = sub[sub["Use Code"] == "SRES"]
```

```python
years = list(set(sub["Closed Roll Year"].values))
useCode = list(set(sub["Use Code"].values))
classCode = list(set(sub["Property Class Code"].values))
```

```python
newS = []
sub["Assessed Improvement Value"] = sub["Assessed Improvement Value"].fillna(0)
for i in sub["Assessed Improvement Value"].values:
    if type(i) == float:
        i = math.floor(i)
        newS += [i]
    if type(i) == str:
        newS += [int(i.replace(',', ''))]
    else:
        newS += [i]
#newS
```

```python
sub["Assessed Improvement Value"] = newS
```

```python
##Fixtures
```

```python
newS1 = []
sub["Assessed Fixtures Value"] = sub["Assessed Fixtures Value"].fillna(0)
for i in sub["Assessed Fixtures Value"].values:
    if type(i) == float:
        i = math.floor(i)
        newS1 += [i]
    if type(i) == str:
        newS1 += [int(i.replace(',', ''))]
    else:
        newS1 += [i]
#newS1
```

```python
sub["Assessed Fixtures Value"] = newS1
```

```python
##Land Value
```

```python
newS2 = []
sub["Assessed Land Value"] = sub["Assessed Land Value"].fillna(0)
for i in sub["Assessed Land Value"].values:
    if type(i) == float:
        i = math.floor(i)
        newS2 += [i]
    if type(i) == str:
        newS2 += [int(i.replace(',', ''))]
    else:
        newS2 += [i]
#newS2
```

```python
sub["Assessed Land Value"] = newS2
```

```python
sub.head()
```

```python
sub["Value"] = sub["Assessed Fixtures Value"] + sub["Assessed Improvement Value"] + sub["Assessed Land Value"]
```

```python
sub.head()
```

```python
subCols = list(sub.columns)
subCols
```

```python
## 5-8] number rooms
##11-15 areas
```

```python
X = sub.iloc[:,[5,6,7,8,11,12,13,14,15]]
X.head()
```

```python
xCols = list(X.columns)
xCols
```

```python
for i in xCols:
    X[i] = X[i].fillna(0)
    newSeries = []
    for j in X[i].values:
        if type(j) == float:
            j = math.floor(j)
            newSeries += [j]
        elif type(j) == str:
            newSeries += [int(j.replace(',', '').replace('.',''))]
        else:
            newSeries += [j]
    X[i] = newSeries
```

```python
[type(X[i][3]) for i in X.columns]
```

```python
X.head()
```

```python
#X = sub.iloc[:,[13]]
#X.head()
```

```python
Y = sub["Value"]
#Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = .2, random_state = 100)
print ('Number of samples in training data:',len(x_train))
print ('Number of samples in validation data:',len(x_test))
```

```python
model = linear_model.LinearRegression().fit(x_train, y_train)
model
```

```python
y_train_pred = model.predict(x_train)
training_error = mse(y_train, y_train_pred)
#(1 - linearReg_model.score(x_train, y_train)) * 100
print("Training Error (Mean squared error) :",training_error)

y_test_pred = model.predict(x_test)
test_error= mse(y_test, y_test_pred)
#(1 - linearReg_model.score(x_test, y_test)) * 100
print("Test Error (Mean squared error) :",test_error)
```
