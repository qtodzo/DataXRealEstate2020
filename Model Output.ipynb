```python
import pickle
import pandas as pd
import numpy as np
```

```python
infile = open('Out', 'rb')
output = pickle.load(infile)
infile.close
```

```python
Projections = pd.DataFrame(output)
```

```python
Projections = Projections.transpose()
Projections.head()
```

```python
Projections.rename(columns = {0:"Year 1", 1:"Year 3", 2:"Year 5", 3:"Year 10", 4:"Year 20"}, inplace = True)
```

```python
def valuation(district, value):
    returns  = Projections.loc[district]
    
    valuationForecast = [value * year for year in returns[:4]]
    return valuationForecast
```
